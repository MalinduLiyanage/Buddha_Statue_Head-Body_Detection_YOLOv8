```python
from PIL import Image
from ultralytics import YOLO
import time

# # Load a pretrained YOLOv8n model
model = YOLO('/content/drive/MyDrive/Colab Notebooks/Object Detector/runs/detect/train3/weights/best_saved_model/best_float32.tflite')

image_name = '/content/drive/MyDrive/Colab Notebooks/Object Segmentation_UNet/20240605_Dataset/test/images/325.jpg'

start = time.time()
results = model(f'{image_name}')  # results list
print(f'run_time:{time.time() - start:.2f}sec')

for r in results:
    im_array = r.plot()  # plot a BGR numpy array of predictions
    im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
    im.show()  # show image
    im.save(f'output/detection.jpg')  # save image

import matplotlib.pyplot as plt

# Load the saved image
saved_image_path = 'output/detection.jpg'
saved_image = Image.open(saved_image_path)

# Plot the image
plt.figure(figsize=(10, 10))
plt.imshow(saved_image)
plt.axis('off')
plt.show()
```
